```python
%sql
CREATE SCHEMA IF NOT EXISTS de_workspace26.gold_aditya_sah_assessment;
```

### Monthly Revenue by Region

```python
%sql
CREATE OR REPLACE VIEW de_workspace26.gold_aditya_sah_assessment.monthly_revenue_by_region AS
SELECT
    YEAR(order_date) AS year,
    MONTH(order_date) AS month,
    region,
    SUM(revenue) AS total_revenue
FROM de_workspace26.silver_aditya_sah_assessment.orders
GROUP BY year, month, region;
```

### Top Products by Category

```python
%sql
CREATE OR REPLACE VIEW de_workspace26.gold_aditya_sah_assessment.top_products AS
SELECT *
FROM (
  SELECT
    category,
    product_name,
    SUM(revenue) AS total_revenue,
    RANK() OVER (PARTITION BY category ORDER BY SUM(revenue) DESC) AS rnk
  FROM de_workspace26.silver_aditya_sah_assessment.orders
  GROUP BY category, product_name
)
WHERE rnk <= 10;
```

```python
%sql
SELECT
    YEAR(order_date) AS year,
    MONTH(order_date) AS month,
    DATE_FORMAT(order_date, 'MMM') AS month_name,
    region,
    SUM(revenue) AS total_revenue
FROM de_workspace26.silver_aditya_sah_assessment.orders
GROUP BY year, month, month_name, region
ORDER BY month;
```
